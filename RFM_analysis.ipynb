### Theory 

> User profile = RFM Modeling (Recency, Frequency, Monetary)

- Recency : how recently a customer has transacted with a brand = **Assign a higher score** to customers who **made a purchase more recently**
> The more recently a customer has made a purchase with a company, the more likely they will continue to keep the business and brand in mind for subsequent purchases. Compared with customers who have not bought from the business in months or even longer periods, the likelihood of engaging in future transactions with recent customers is arguably higher.
- Frequency : how frequently they’ve engaged with a brand = **Assign a higher score** to customers who **make more purchases (transactions) in a specific time**
> The frequency of a customer’s transactions may be affected by factors such as the type of product, the price point for the purchase, and the need for replenishment or replacement. 
> If the purchase cycle can be predicted—for example, when a customer needs to buy more groceries—marketing efforts may be directed toward reminding them to visit the business when staple items run low.
- Monetary : how much money they’ve spent on a brand’s products and services = **Assign a higher score** to customers who **spend more money**.
> Monetary value stems from how much the customer spends. A natural inclination is to put more emphasis on encouraging customers who spend the most money to continue to do so. 
> While this can produce a better return on investment (ROI) in marketing and customer service, it also runs the risk of alienating customers who have been consistent but may not spend as much with each transaction.

## 6. Data Processing - RFM calculating 

```python
# Recency : how recently a customer has transacted with a brand = **Assign a higher score** to customers who **made a purchase more recently**
```

```python
ecm['Cost'] = ecm['Quantity'] * ecm['UnitPrice']
last_day = ecm['Date'].max() 

RFM_ecm = ecm.groupby('CustomerID').agg(
    Recency = ('Date', lambda x: last_day - x.max()), 
    Frequency = ('CustomerID', 'count'), 
    Monetary = ('Cost', 'sum') 
).reset_index()

RFM_ecm['Recency'] = RFM_ecm['Recency'].dt.days.astype('int16')
RFM_ecm.dtypes
```

```python
RFM_ecm
```

```python
RFM_ecm['R'] = pd.qcut(RFM_ecm['Recency'], 5, labels= range(1,6)).astype(str)
RFM_ecm['F'] = pd.qcut(RFM_ecm['Frequency'], 5, labels= range(1,6)).astype(str)
RFM_ecm['M'] = pd.qcut(RFM_ecm['Monetary'], 5, labels= range(1,6)).astype(str)
RFM_ecm['RFM'] = RFM_ecm.apply(lambda x: x['R'] + x['F'] + x['M'], axis= 1)
```

```python
RFM_ecm
```

```python
seg = pd.read_excel('ecommerce_retail.xlsx', sheet_name = 'Segmentation')
seg['RFM Score'] = seg['RFM Score'].str.split(',')
seg = seg.explode('RFM Score').reset_index(drop = True)
```

```python
seg 
```

```python
# loai bo khoang trang giua cac so sau khi split 
seg['RFM Score'] = seg['RFM Score'].apply(lambda x: x.replace(" ", ""))
```

```python
# merge 2 bang 
ecm_f = RFM_ecm.merge(seg, left_on= 'RFM', right_on='RFM Score', how='left')
ecm_f
```

```python
ecm_distribution_table = ecm_f[['Segment','CustomerID','Recency','Frequency','Monetary']].groupby(['Segment']).agg({'CustomerID':'count',
                                                                                                                    'Monetary':'sum', 
                                                                                                                    'Recency':'mean',
                                                                                                                    'Frequency':'mean'}).reset_index().rename(columns={'CustomerID': 'user_volume',
                                                                                                                                                                       'Monetary':'total_spent', 
                                                                                                                                                                       'Recency': 'avg_recc',
                                                                                                                                                                       'Frequency': 'avg_freq'}).round(1)
ecm_distribution_table['avg_spent'] = round(ecm_distribution_table['total_spent'] / ecm_distribution_table['user_volume'], 2)

# cách này không tạo barplot được 

ecm_index = ecm_distribution_table.set_index('Segment')
pivot_ecm = ecm_index.T
pivot_ecm
```

```python
# tạo bảng phù hợp để tạo barplot 
segment_by_user_count = ecm_f[['Segment','CustomerID']].groupby(['Segment']).agg({'CustomerID':'count'}).reset_index().rename(columns={'CustomerID': 'volume'})
segment_by_user_count['distribution_percent'] = round((segment_by_user_count['volume'] / segment_by_user_count['volume'].sum() * 100), ndigits= 2)
segment_by_user_count['type'] = 'volumn distribution'

segment_by_user_spent = ecm_f[['Segment','Monetary']].groupby(['Segment']).agg({'Monetary':'sum'}).reset_index().rename(columns={'Monetary': 'volume'})
segment_by_user_spent['distribution_percent'] = round((segment_by_user_spent['volume'] / segment_by_user_spent['volume'].sum() * 100), ndigits= 2)
segment_by_user_spent['type'] = 'spent distribution' 

ecm_distribution = pd.concat([segment_by_user_count, segment_by_user_spent], ignore_index=True)
```

```python
order_ = ['Champions', 'Loyal', 'Potential Loyalist', 'New Customers', 'Promising', 'Need Attention', 'About To Sleep', 'At Risk', 'Cannot Lose Them', 'Hibernating customers', 'Lost customers']
ecm_distribution_table['Segment'] = pd.Categorical(ecm_distribution_table['Segment'], categories= order_, ordered= True)
ecm_distribution_table = ecm_distribution_table.sort_values('Segment')
ecm_distribution_table
```
